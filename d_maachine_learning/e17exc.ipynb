```python
from sklearn.datasets import load_digits
digits = load_digits()
dir(digits)
```

```python
import pandas as pd 
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df
```

```python
x = df[['pixel_0_0','pixel_0_1','pixel_0_2','pixel_0_3','pixel_0_4','pixel_0_5','pixel_0_6','pixel_0_7']]
y = digits.target
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from cProfile import label
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.scatter(x.pixel_0_0, y, label='pixel_0_0')
plt.scatter(x.pixel_0_1, y, label='pixel_0_1')
plt.scatter(x.pixel_0_2, y, label='pixel_0_2')
plt.scatter(x.pixel_0_3, y, label='pixel_0_3')
plt.scatter(x.pixel_0_4, y, label='pixel_0_4')
plt.scatter(x.pixel_0_5, y, label='pixel_0_5')
plt.scatter(x.pixel_0_6, y, label='pixel_0_6')
plt.scatter(x.pixel_0_7, y, label='pixel_0_7')

plt.legend()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pre = knn.predict(x_test)

truth = y_test
prediction = y_pre

cm = confusion_matrix(truth, prediction)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn

plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('prediction')
plt.ylabel('truth')
```
