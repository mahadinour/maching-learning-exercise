```python
df = pd.read_csv("a5weather_data.csv", parse_dates=["day"])
df.set_index('day',inplace=True)
df
```

fillna --> fill missing value

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
	'windspeed' :0,
    'event' :'No Event'
})
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="ffill", axis='columns')
new_df
```

```python
df.fillna(method='ffill', limit=1)
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time")
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```
