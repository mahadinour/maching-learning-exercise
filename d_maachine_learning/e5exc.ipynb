```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('e5carprices.csv')
df
```

```python
dummy = pd.get_dummies(df['Car Model'])
dummy
```

```python
df = df.drop('Car Model',axis=1)
df
```

```python
marged = pd.concat([df, dummy], axis='columns')
marged
```

```python
reg = linear_model
```

```python
x = marged[['Mileage', 'Age(yrs)','Audi A5', 'BMW X5', 'Mercedez Benz C class']]
x
```

```python
y = marged['Sell Price($)']
y
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x, y)
# x = 'Mileage', 'Age(yrs)','Audi A5', 'BMW X5', 'Mercedez Benz C class'
# y = 'Sell Price($)'
```

```python
reg.score(x,y)
```
