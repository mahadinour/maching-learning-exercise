```python
import pandas as pd
```

```python
df = pd.read_csv('e14titanic.csv')
df= df.drop(df[['Name', 'Ticket','Cabin', 'Embarked']], axis='columns')
df
```

divided into input and target

```python
input = df.drop('Survived', axis='columns')
target= df['Survived']
input
```

```python
dummy = pd.get_dummies(input["Sex"])
dummy.head(2)
```

```python
input = pd.concat([input, dummy], axis='columns')
input.drop(['Sex', 'female'], axis='columns', inplace=True)
input
```

```python
input.info()
```

fill the null value

```python
input.Age = input.Age.fillna(input.Age.interpolate())
```

```python
from sklearn.model_selection import train_test_split
x = input
y = target
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
