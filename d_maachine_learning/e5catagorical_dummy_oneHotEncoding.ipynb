```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('e5homeprices.csv')
df
```

```python
dummy = pd.get_dummies(df['town'])
dummy
```

```python
marged = pd.concat([df,dummy], axis='columns')
marged
```

```python
final_marged = marged.drop(['town','west windsor' ], axis='columns')
final_marged
```

```python
reg = linear_model.LinearRegression()
```
