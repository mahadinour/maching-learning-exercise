```python
import matplotlib.pyplot as plt 
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins=3)
```

```python
plt.hist(blood_sugar, bins=[80,100, 125,150], rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95, color='g', label='suger')
plt.legend()
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95, color='g', histtype='step')
```

```python
blood_sugar_man = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
#plt.hist(blood_sugar, bins=3, rwidth=0.95, color='g', label='suger')
plt.hist([blood_sugar_man, blood_sugar_women], bins=[80,90,100,110,120,130,140,150], rwidth=0.95, color=['green', 'orange'], label=['suger_man', 'suger_weman'])
plt.legend()
```

```python
plt.xlabel('suger lable')
plt.ylabel('num. of pacient')
plt.hist([blood_sugar_man, blood_sugar_women], bins=[75,100,125,150], rwidth=0.95, color=['green', 'orange'], label=['suger_man', 'suger_weman'])
plt.legend()
```

```python
plt.xlabel('suger lable')
plt.ylabel('num. of pacient')
plt.hist([blood_sugar_man, blood_sugar_women], bins=[80,90,100,110,120,130,140,150], rwidth=0.8, color=['green', 'orange'], label=['suger_man', 'suger_weman'])
plt.legend()
plt.grid()
```
