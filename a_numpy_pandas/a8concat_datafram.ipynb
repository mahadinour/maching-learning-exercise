```python
import pandas as pd
```

```python
bangladesh_wather = pd.DataFrame({
    'city' : ['dhdka', 'comilla', 'sylhet'],
    'temparature' : [32, 45, 50],
    'himidity' : [80, 60, 75]
})
bangladesh_wather
```

```python
india_wather = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglor'],
    'temparature' : [35, 40,20],
    'himidity' : [60,70, 56]
})
india_wather
```

```python
usa_wather = pd.DataFrame({
    'city' : ['NY', 'taxus', 'sunfrancisco'],
    'temparature' : [32, 45, 50],
    'himidity' : [80, 60, 75]
})
usa_wather
```

```python
df = pd.concat([bangladesh_wather, india_wather, usa_wather])
df
```

```python
df = pd.concat([bangladesh_wather, india_wather, usa_wather], ignore_index=True)
df
```

```python
df = pd.concat([bangladesh_wather, india_wather, usa_wather],keys=['bangladesh','india', 'usa'])
df
```
